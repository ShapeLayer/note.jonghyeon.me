```python
import matplotlib.pyplot as plt
import numpy as np

p_success_single_trial = 1e-4
p_fail_single_trial = 1 - p_success_single_trial

# trials
X = np.logspace(2, 5, 200)

prob_fail_all_trials = p_fail_single_trial ** X

# prob_success_at_least_once
y = 1 - prob_fail_all_trials


plt.figure(figsize=(12, 7))


threshold = 0.7
trials_at_threshold = np.log(1 - threshold) / np.log(1 - p_success_single_trial)
plt.axhline(y=threshold, color='r', linestyle='--', linewidth=2, label=f'Passed least once at {threshold:.2f} probability')
plt.annotate(
    f'~{trials_at_threshold:.0f} trials',
    xy=(trials_at_threshold, threshold),
    xytext=(trials_at_threshold * 1.1, threshold + 0.05),
    arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2", color='black'),
    fontsize=10,
    ha='left',
    va='bottom'
)


plt.plot(X, y, label=f'P(Success at least once) for p={p_success_single_trial}')

plt.xlabel('trials', fontsize=12)
plt.ylabel('success at least once', fontsize=12)
plt.xscale('log')
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.legend(fontsize=10)
plt.tight_layout()
plt.show()
```
